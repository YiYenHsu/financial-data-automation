### **解壓縮資料夾內的<font color="red">所有</font>檔案**

```python
import os
import zipfile
from datetime import datetime

def unzip_files(zip_files, destination_root):
    """
    解壓縮檔案到根據壓縮檔名稱建立的新資料夾。

    :param zip_files: 壓縮檔案的完整路徑列表
    :param destination_root: 解壓縮檔案的根目錄
    """
    for zip_file in zip_files:
        if not zipfile.is_zipfile(zip_file):
            print(f"{zip_file} 不是有效的壓縮檔案，跳過。")
            continue

        # 獲取壓縮檔案名稱（不含副檔名）
        folder_name = os.path.splitext(os.path.basename(zip_file))[0]
        target_folder = os.path.join(destination_root, folder_name)

        # 如果資料夾不存在則建立
        os.makedirs(target_folder, exist_ok=True)

        # 解壓縮檔案
        with zipfile.ZipFile(zip_file, 'r') as zip_ref:
            zip_ref.extractall(target_folder)

        print(f"已解壓縮 {zip_file} 到 {target_folder}")

# 範例使用
if __name__ == "__main__":
    # 動態列出所有 "新增報告" 的壓縮檔案
    source_folder = "./data/raw/zip_temp/"
    zip_files = [
        os.path.join(source_folder, file)
        for file in os.listdir(source_folder)
        if file.endswith(".zip")
    ]

    # 設定解壓縮的根目錄
    destination_root = "./data/processed/daily_broker_reports/"

    # 呼叫函式
    unzip_files(zip_files, destination_root)

result = datetime.now().strftime("%Y-%m-%d %H:%M:%S %p")
print(result)
print("完成")
```

### **解壓縮資料夾內的<font color="red">指定</font>檔案**

```python
import os
import zipfile
from datetime import datetime

def unzip_files(zip_files, destination_root):
    """
    解壓縮檔案到根據壓縮檔名稱建立的新資料夾。

    :param zip_files: 壓縮檔案的完整路徑列表
    :param destination_root: 解壓縮檔案的根目錄
    """
    while zip_files:
        for zip_file in zip_files:
            if not zipfile.is_zipfile(zip_file):
                print(f"{zip_file} 不是有效的壓縮檔案，跳過。")
                continue

            # 獲取壓縮檔案名稱（不含副檔名）
            folder_name = os.path.splitext(os.path.basename(zip_file))[0]
            target_folder = os.path.join(destination_root, folder_name)

            # 如果資料夾不存在則建立
            os.makedirs(target_folder, exist_ok=True)

            # 解壓縮檔案
            with zipfile.ZipFile(zip_file, 'r') as zip_ref:
                zip_ref.extractall(target_folder)

            print(f"已解壓縮 {zip_file} 到 {target_folder}")

            # 解壓縮後移除檔案
            zip_files.remove(zip_file)

# 範例使用
if __name__ == "__main__":
    # 動態列出資料夾內所有的壓縮檔案，僅選取 20231214 開始的檔案
    source_folder = "./data/raw/zip_temp/"
    zip_files = [
        os.path.join(source_folder, file)
        for file in os.listdir(source_folder)
        if file.endswith(".zip") and file >= "20240607.zip"
    ]

    # 設定解壓縮的根目錄
    destination_root = "./data/processed/daily_broker_reports/"

    # 呼叫函式
    unzip_files(zip_files, destination_root)


result = datetime.now().strftime("%Y-%m-%d %H:%M:%S %p")
print(result)
print("完成")
```
