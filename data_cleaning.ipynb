### 第二版:year&season改成變量

```python
import os
import pandas as pd
import glob
import re

#year = 108
#season = 1

#CSV檔的檔案資料夾 找裡面的全部檔案路徑
dirPath = rf"D:\Income_STMT\*.csv"
listname = glob.glob(dirPath)

df = pd.DataFrame(listname, columns=['代號'])
file_name = "損益表.xlsx"
df.to_excel(file_name, index=False, )
print(file_name + '匯出完成')

#匯入檔案路徑，轉成dataframe
excel_file_path = "損益表.xlsx"
df = pd.read_excel(excel_file_path, sheet_name='Sheet1')
csv_files = df['代號']
#print(csv_files)

#帶入迴圈，轉成xlsx檔
for csv_file in csv_files:
    
    df = pd.read_csv(csv_file)
    
    stock_code = csv_file[19:23]
    year = csv_file[15:18]

    excel_file_name = str(year) + "_" + str(stock_code) + "_" + "損益表.xlsx"
    df.to_excel(excel_file_name, sheet_name='Sheet1', index=False)
    
    print(excel_file_name)
    
print(f"{file_name}轉檔完成")
```

### 第一版

```python
import os
import pandas as pd
import glob
import re

year = 112
season = 3

dirPath = r"D:\BalanceSheet\112\Q3\*.csv"
listname = glob.glob(dirPath)

df = pd.DataFrame(listname, columns=['代號'])
file_name = "112Q3.xlsx"
df.to_excel(file_name, index=False, )
print(file_name + '匯出完成')

excel_file_path = '112Q3.xlsx'
df = pd.read_excel(excel_file_path, sheet_name='Sheet1')

csv_files = df['代號']
#print(csv_files)

for csv_file in csv_files:
    
    df = pd.read_csv(csv_file)
    
    stock_code = csv_file[29:33]

    excel_file_name = str(year) + "Q" + str(season) + "_" + str(stock_code) + "_" + "資產負債表.xlsx"
    df.to_excel(excel_file_name, sheet_name='Sheet1', index=False)
    
    print(excel_file_name)
    
print({year} + {season} + '轉檔完成')
```
